```python
city = pd.DataFrame(narrow_city_data_df["City"])
country = pd.DataFrame(narrow_city_data_df["Country"])
humidity = pd.DataFrame(narrow_city_data_df["Humidity"])
lng= pd.DataFrame(narrow_city_data_df["Lng"])
lat = pd.DataFrame(narrow_city_data_df["Lat"])
```

```python
hotel_df = pd.concat([city, country, humidity, lng, lat], axis=1)
hotel_df
```

```python
base_url = "https://api.geoapify.com/v2/places"
radius = 10000
for index, row in hotel_df.iterrows():
    city = hotel_df.loc[index, "City"]
    lat = hotel_df.loc[index, "Lat"]
    lng = hotel_df.loc[index, "Lng"]
    
    filters = f"circle:{lng},{lat},{radius}"
    bias = {f"proximity:{lng},{lat}"}
    categories = "accommodation.hotel"
    
    params = {
        "filter": filters,
        "bias": bias,
        "categories": categories,
        "apiKey":geoapify_key
    }
    
    try: 
        hotel = requests.get(base_url, params=params)
        hotel = hotel.json()
        hotel_df.loc[index, "Hotel Name"] = hotel["features"][0]["properties"]["name"]
    except (KeyError, IndexError):
        # If no restaurant is found, set the restaurant name as "No restaurant found".
        hotel_df.loc[index, "Hotel Name"] = "No restaurant found"

# Display sample data
hotel_df
    
```

```python
hotel_data = hotel_df.hvplot.points(
    "Lng",
    "Lat",
    geo = True,
    tiles = "OSM",
    frame_width = 700,
    frame_height = 500,
    size = "Humidity",
    color = "City",
    hover_cols = ["Lng", "Lat", "City", "Humidity", "Hotel Name", "Country"]
)

# Display the map plot_1
hotel_data
```
